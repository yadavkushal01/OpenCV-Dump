### **Face and Eye Detection using HAAR Cascade Classifiers**

##### HAAR Cascade Classifier

It helps us to detect object and is an **effective object detection method**. It is a **machine learning based approach** where a cascade function is **trained** from a lot of positive and negative images. It is then used to **detect objects in other images.**

Even a 24x24 window results over 160000 features .how do we select the best features out of 160000+ features? It is achieved by **Adaboost** 

**ADABOOST** --> For each feature, it finds the best threshold which will classify the faces to positive and negative. Obviously, there will be errors or misclassifications. We select the features with minimum error rate. For this the same process is done multiple times. New error rates are calculated. Also new weights. The process is continued until the required accuracy or error rate is achieved or the required number of features are found. So now you take an image. Take each 24x24 window. Apply 6000 features to it. Check if it is face or not. Wow.. Isn't it a little inefficient and time consuming?

the concept of **Cascade of Classifiers**. Instead of applying all 6000 features on a window, the features are grouped into different stages of classifiers and applied one-by-one. 6000+ features with 38 stages with 1, 10, 25, 25 and 50 features in the first five stages. (The two features in the above image are actually obtained as the best two features from Adaboost)

OpenCV provides a training method (see Cascade Classifier Training) or pretrained models, that can be read using the CascadeClassifier::load method. The pretrained models are located in the data folder in the OpenCV installation.

**First, a cv::CascadeClassifier is created and the necessary XML file is loaded using the cv::CascadeClassifier::load method. Afterwards, the detection is done using the cv::CascadeClassifier::detectMultiScale method, which returns boundary rectangles for the detected faces or eyes.**

This include various features within a particular phase like : 

**Edge features , Line features , Four-Rectangle features**   -->  these features are stored in some of XML files

```python
import numpy as np
```

to detect face in image

```python
# we point OpenCV's CascadeClassifier inbuilt function to where we stored our classifiers ( XML files ) and it loads it

face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')  # haarcascade classifier to recognise using frontal face of a person
# face_classifier is object to determine the face in a particular image 

# load our image then covert it to grayscale --> converting from BGR color/channel (Blue,Green,Red) to Gray color/channel
image = cv2.imread('./images/Obama.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # cvtColor to convert image to other colors which take image object as argument 
# the format of storing image COLOR_BGR2GRAY in this case where BGR2GRAY is converting blue, green, red to gray
# to increase speed of image processing we change it to GRAY channel

# Our classifier returns the ROI of the detected face as a tuple 
# It stores the top left coordinate and the bottom right coordinates
faces = face_classifier.detectMultiScale(gray,1.3,5)   # this function of haarcascade classifier helps us to findout the location of our face and height and width of the face 
# this gives 4 values including ( x-coordinate, y-coordinate, width size, height size ) of the face in that image 
# it takes image, scale-factor and minNeighbour (gray,1.3,5)

if len(faces)==0: # provides empty 
    print('No faces found')

# Now we get the faces in faces object and there can be multiple faces so we use loop to detect and draw a rectangle around every face it found in faces tuple

# We iterate through our faces array and draw a rectangle
# over each face in faces
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w,y+h), (255,0,0) ,2)  # takes arguments of starting coordinates where the rectangle is to be made (x,y) and end coordinate (x+w,y+h)
    # (127,0,255) is for the color of rectangle (pink)
    cv2.imshow('Face Detection',image)
    cv2.waitKey()

cv2.destroyAllWindows()


# print(type(faces))
# print(faces)
```

combining face and eye detection

there maybe exceptions with some of images in detection

```python
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_classifier=cv2.CascadeClassifier('haarcascade_eye.xml')

image=cv2.imread('./images/Trump.jpg')
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

faces=face_classifier.detectMultiScale(gray,1.3,5)

if len(faces)==0:
    print('No Face Found!')

for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(127,0,255),2)
    cv2.imshow(mat=image,winname='Face and Eye Detection')
    cv2.waitKey()

    # till here it is same to draw rectangle when face found

    roi_gray=gray[y:y+h,x:x+w]  # extracting the rectangular part where face found from the grey image
    roi_color=image[y:y+h,x:x+w]   # doing same with colored image 
    eyes=eye_classifier.detectMultiScale(roi_gray)  # given gray face image for processing 
    for (ex,ey,ew,eh) in eyes: # accepted all parameter
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,255,0),2)  # make rectangle on where eye found
        cv2.imshow('image',image) # shows image
        cv2.waitKey()

cv2.destroyAllWindows()
```

### Face and Eye Detection from Videos using haarcascade Classifiers

```python
# Face Recognition 

face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade=cv2.CascadeClassifier('haarcascade_eye.xml')

def detect(gray,frame):
    faces=face_cascade.detectMultiScale(gray,1.3,5)   # takes faces from that frame of videos(image) and stores in faces(array)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)  # draw rectangle around where face found
        roi_gray=gray[y:y+h,x:x+w] # takes out that rectangle part and stores in roi_gray for gray frame
        roi_color=frame[y:y+h,x:x+w] # from original colored frame rectangle taken out
        eyes=eye_cascade.detectMultiScale(roi_gray)  # eyes are detected and stored in eyes(array)
        for (ex,ey,ew,eh) in eyes: 
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)  # now rectangle made out where eye found 
    return frame # this function returns that particular frame

# Doing some Face Recognition with webcam

video_capture=cv2.VideoCapture(0)  # used to access webcam from device and opens it when called and capture frames in seconds
while True:
    _,frame=video_capture.read() # reading the frames captured by video_capture from webcame  --> '_,frame' is to read all the frames in one time for first frame we use 'ret,frame'
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)  # Converting frame to Gray color
    canvas=detect(gray,frame)  # calling the above made function and storing the processed image/frame in canvas object
    cv2.imshow('Video',canvas)  # showing that particular frame that is detected to window
    if cv2.waitKey(1) & 0xFF == ord('k'):  # to quilt the while loop
        break
video_capture.release()  # release all the object we created using video_capture (webcam)
cv2.destroyAllWindows() 

```
