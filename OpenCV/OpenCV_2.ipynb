Car and Pedistrain detection/

Pedistrain Detection :

```python
import cv2 
import numpy as np

# Creating body classifier
body_classifier=cv2.CascadeClassifier('haarcascade_fullbody.xml') # cascade for detecting the whole body movement

#Initiating video capture for video file
vid_cap=cv2.VideoCapture('./images/image_examples_walking.avi') # capturing the video in an object
#loop once video is succesfully loaded
while vid_cap.isOpened():  # if video is not ended or still running
    #Read first frame
    ret,frame=vid_cap.read()  # construction of frames from video object
    # frame=cv2.resize(frame,None,fx=1,fy=1,interpolation=cv2.INTER_LINEAR)

    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) #convert frame to grey for processing
    #pass frame to our body classifier
    bodies=body_classifier.detectMultiScale(gray,1.2,3) # detect the bodies in frame

    #extract bounding boxes for any bodies identified
    for (x,y,w,h) in bodies:
        cv2.rectangle(frame,(x,y),(x+w,y+w),(0,255,255),2) #draw rectangular shape around body
        cv2.imshow('Pedestrains',frame)
    
    if cv2.waitKey(1) == "k": # k is the Enter key
        break

vid_cap.release()
cv2.destroyAllWindows()
```

Cars Detection : 

```python
import cv2
import numpy as np
import time

# Create our body classifier
car_classifier=cv2.CascadeClassifier('haarcascade_car.xml')

# Initiate video capture for video file
vid_cap=cv2.VideoCapture('./images/image_examples_cars.avi')

# Loop once video is successfully loaded
while vid_cap.isOpened():
    time.sleep(.05)
    #Read first frame
    ret,frame=vid_cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    cars=car_classifier.detectMultiScale(gray,1.4,2) # frame where data is loaded and scaleFactor is 1.4 and min neighbour is 2

    # Pass frame to our car classifier 
    for (x,y,w,h) in cars:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        # cv2.circle(frame,(x+w,y+h),30,(255,0,0),2)  # trying to use circle as representation
        cv2.imshow('Car Detector',frame)

    if cv2.waitKey(1) == 13:
        break

vid_cap.release()
cv2.destroyAllWindows()
```
